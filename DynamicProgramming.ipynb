 Memoization without lru_cache

```python
def square(x):
    return x**2

import time
import random

arr = [random.randint(1,8) for _ in range(100000)]
t1 = time.time()
for i in arr:
    print(square(i))
t2 = time.time()
```

```python
print(t2 - t1)
```

```python
cache={}
def squarenew(x):
    if x in cache:
        return cache[x]
    else:
        cache[x] = x**2
        return cache[x]
```

```python
t1 = time.time()
for i in arr:
    print(squarenew(i))
t2 = time.time()
```

```python
print(t2 - t1)
```

```python
from functools import lru_cache

@lru_cache(maxsize=10000)
def squaring(x):
    return x**2

```

```python
t1 = time.time()
for i in arr:
    print(squaring(i))
t2 = time.time()
```

```python
print(t2 - t1)
```
